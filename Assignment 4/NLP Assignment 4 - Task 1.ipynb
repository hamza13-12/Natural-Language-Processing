```python
import pandas as pd
from gensim.models import Word2Vec
import gensim.downloader as api
```

```python
df = pd.read_csv(r"C:\Users\Hamza\Desktop\NLP\Assignment 4\roman_urdu_text.csv", delimiter='\t')
```

```python
df.head(10)
```

```python
df_cleaned = df.dropna(subset=['smsreceived'])
```

```python
df_cleaned.head(10)
```

```python
texts = df_cleaned['smsreceived'].astype(str).tolist()
```

```python
texts[0] #Sanity check
```

```python
tokenized_texts = [text.split() for text in texts]
```

```python
model = Word2Vec(sentences=tokenized_texts, vector_size=300, window=5, min_count=1, workers=4)
```

```python
model
```

```python
similar_words = model.wv.most_similar("police", topn=10)
print("Top 10 words most similar to 'police':")
for word, similarity in similar_words:
    print(f"{word}: {similarity}")
```

```python
model.wv.save_word2vec_format('word2vec_embeddings.txt', binary=False)
```
